```python
# Step 2: Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Step 3: Load dataset after upload
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
# Step 4: See first 5 rows of the dataset
data.head()
```

```python
# Step 5: Scatter plot - Age vs Purchased
plt.scatter(data.Age, data.Purchased)
plt.xlabel('Age')
plt.ylabel('Purchased')
plt.show()
```

```python
# Step 6: Scatter plot - EstimatedSalary vs Purchased
plt.scatter(data.EstimatedSalary, data.Purchased)
plt.xlabel('Estimated Salary')
plt.ylabel('Purchased')
plt.show()
```

```python
# Step 7: Select features columns
x = data[["Age", "EstimatedSalary"]]
x
```

```python
# Step 8: Select target column
y = data['Purchased']
y
```

```python
# Step 9: Divide dataset into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
# Step 10: Check test data features
x_test
```

```python
# Step 11: Build logistic regression model
model = LogisticRegression()
```

```python
# Step 12: Train the model
model.fit(x_train, y_train)
```

```python
# Step 13: Get predictions from the model on test data
model.predict(x_test)
```

```python
# Step 14: Check model accuracy
model.score(x_test, y_test)
```

```python
# Step 15: Prepare new data to predict
new_values = np.array([[20, 10000], [25, 15000], [40, 50000]])
feature_names = ["Age", "EstimatedSalary"]
new_data = pd.DataFrame(new_values, columns=feature_names)
new_data
```

```python
# Step 16: Predict new data results
new_predictions = model.predict(new_data)
new_predictions
```
